```python
import pyhmmer
from Bio import SeqIO
from Bio import AlignIO
pyhmmer.__version__

align = AlignIO.read("alignment.sto", "stockholm")
print(align)
```

```python
#create amino acid alphabet
alphabet = pyhmmer.easel.Alphabet.amino()
```

```python
#read in the msa
with pyhmmer.easel.MSAFile("alignment.sto", digital=True, alphabet=alphabet) as msa_file:
    msa = msa_file.read()
    print(msa)
```

```python
#this entire function build the HMM
#important note, inside the created "alignment.sto" file you need on the second line 
#to add a name for the pyhmmer package to create an HMM e.g : #=GF ID TEM-1
#1.# STOCKHOLM 1.0
#2.#=GF ID TEM-1 (this is the line)
#3.#=GF SQ 10
builder = pyhmmer.plan7.Builder(alphabet)
background = pyhmmer.plan7.Background(alphabet)
hmm, _, _ = builder.build_msa(msa, background)
```

```python
#just to see the consensus sequence of the HMM
hmm.consensus
```

```python
#save the HMM file
with open("TEM-1.hmm", "wb") as output_file:
    hmm.write(output_file)
```

```python
import pyhmmer
#run and compare the query sequence to the HMM
pipeline = pyhmmer.plan7.Pipeline(alphabet, background=background)
with pyhmmer.easel.SequenceFile("sequence.842.fasta", digital=True, alphabet=alphabet) as seq_file:
    hits = pipeline.search_hmm(hmm, seq_file)
```

```python
combined_df = numberingdf.join(numberingdf2)
counter = 0
counternew = 0

def renumber_index(row):
    global counter
    global counternew
    new_index = 0
    if row['standard numbering'] == '-':
        counternew += 1
        new_index = f"{counter}.{counternew}"
    elif row['standard numbering'] != '-':
        counter += 1 
        counternew = 0
        new_index = counter
    return new_index

combined_df['new_index'] = combined_df.apply(renumber_index, axis=1)

print(combined_df.to_string(index=False))
```
